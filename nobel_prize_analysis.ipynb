```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
nob_p = pd.read_csv('nobel.csv')
nob_p.head()
```

```python
# Get the shape of the Dataset
rows, columns = nob_p.shape

# Print the message
print(f"This dataframe contains {rows} rows and {columns} columns")
```

```python
nob_p.info()
```

#### Column names description

1. `year` - Year the Nobel Prize was awarded

2. `category` - Category of the Nobel Prize 

3. `prize` - Nobel Prize title

4. `motivation` - The Reason why the Nobel Prize was awarded

5. `prize_share` - Dictation by Fraction of the Nobel Prize awarded

6. `laureate_id` - Laureate's unique id

7. `laureate_type` - The Type of laureate (Individual / Organization)

8. `full_name` - Award winner's Full name

9. `birth_date` - Award winner's Date of Birth

10. `birth_city` - Award winner's Birth city

11. `birth_country` - Award winner's Birth country

12. `sex` - Award winner's Gender 

13. `organization_name` - The name of the organization associated with the Award winner

14. `organization_city` - Organization's city location

15. `organization_country` - Organization's country location

16. `death_date` - Award winner's Date of Death

17. `death_city` - Award winner's Death city

18. `death_country` - Award winner's Death country

# DATA CLEANING PROCESS

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

```python
# We can extract the year in the birth_date and death_date to get age and lifespan of awardees

nob_p['birth_year'] = nob_p['birth_date'].str.split('-', expand = True)[0]
nob_p['death_year'] = nob_p['death_date'].str.split('-', expand = True)[0]
nob_p.head()
```

```python
# Drop unwanted columns (birth_date, death_date, and laureate_id)

nob_p.drop(columns = ['birth_date', 'death_date', 'laureate_id'], inplace = True)
nob_p.head()
```

```python
nob_p.info()
```

```python
# Change the data-type of wanted columns (birth_year and death_year)
# First to float

nob_p['birth_year'] = pd.to_numeric(nob_p['birth_year'], errors='coerce')
nob_p['death_year'] = pd.to_numeric(nob_p['death_year'], errors='coerce')

# Replace 'NaN' strings with np.nan
nob_p['birth_year'] = nob_p['birth_year'].replace('NaN', np.nan)
nob_p['death_year'] = nob_p['death_year'].replace('NaN', np.nan)

# Then to integer

nob_p['birth_year'] = nob_p['birth_year'].astype('Int64')
nob_p['death_year'] = nob_p['death_year'].astype('Int64')

nob_p.info()
```

```python
# To get age and lifespan of awardees by creating new wanted columns (laureate_age (age at the award year) and laureate_lifespan (age before death))

nob_p['laureate_award_age'] = nob_p['year'] - nob_p['birth_year']

nob_p['laureate_lifespan'] = nob_p['death_year'] - nob_p['birth_year']

nob_p.head()
```

```python
# Drop unwanted columns (birth_year, death_year)

nob_p.drop(columns = ['birth_year', 'birth_city', 'organization_city', 'death_city', 'death_country'], inplace = True)
nob_p.head()
```

## 1) How many Nobel Prizes have been awarded each year?

```python
# NOTE!!! - These results are based on award category selection not number of persons involved (prize share count)

awards_per_year = nob_p.groupby('year')['category'].nunique()
awards_per_year =  awards_per_year.reset_index()
awards_per_year.rename(columns = {'category' : 'nobel prize(s)'}, inplace = True)
awards_per_year
```

## 2) Are there any trends or patterns in the number of prizes awarded over time? 

```python
# NOTE!!! - These trends are based on category selection (Nobel Prizes awarded each year)

plt.figure(figsize=(12, 10))
sns.set(style="whitegrid")

sns.lineplot(data = awards_per_year, x = 'year', y = 'nobel prize(s)', color='b')

plt.annotate('Lowest prize won',
             xy=(1916, 1),
             xytext=(1940, 2.1), fontsize = 14, color = 'green',
             arrowprops=dict(arrowstyle='->', color='black'))

plt.annotate('Stable prize award range', 
             xy=(2022, 5.75), 
             xytext=(1980, 5.75), fontsize = 14, color = 'green')

plt.annotate('', 
             xy=(2023, 5.9), 
             xytext=(1973, 5.9),
             arrowprops=dict(arrowstyle='<->', color='black'))


plt.title('Distribution of Nobel Prizes Won OverTime', fontsize=16, color = 'purple')
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Prizes Won', fontsize=12)
plt.show()
```

###### Answer

###### After Year 1972, Awards won per category has always been 6 `(Check - Nobel Prizes Won OverTime image above (Stable prize award range))`

###### In the Year 1916, Only one laureate award was won. `(Check - Nobel Prizes Won OverTime image above (Lowest prize won))`

## 14) How often are Nobel Prizes shared, and what is the typical share distribution?

```python
# How often are the nobel prizes shared? 
# To answer this, we need to group by year and category

np_sharing_ratio = nob_p.groupby(['year', 'category'])['prize'].count().reset_index()
np_sharing_ratio
```

```python
# Then rename wanted column prize to give insights

np_sharing_ratio_mapping = {1: '1-person award', 2: '2-person award', 3: '3-person award'}

np_sharing_ratio['prize'] = np_sharing_ratio['prize'].replace(np_sharing_ratio_mapping)

np_sharing_ratio.rename(columns={'prize': 'award_portion'}, inplace=True)

np_sharing_ratio
```

```python
# This means a nobel prize can be awarded entirely to one person, divided equally between two people, or shared among three people.

np_sharing_ratio['award_portion'].unique()
```

```python
plt.figure(figsize=(10, 7))
np_sharing_ratio['award_portion'].value_counts().plot(kind = 'pie',
                                              autopct = '%1.0f%%',
                                              colors = ['#6A5ACD', '#3CB371', '#FF6347'],
                                              explode = (0, 0, 0.05), fontsize = 15)
plt.ylabel('')
plt.title('Percentage proportion of award won and shared between laureates (winners) Overtime', fontsize = 15, color = 'purple')
plt.show()
```

```python
# What is the typical share distribution?
# Merging dataframes to create a new one with award_portion

new_nob_p = pd.merge(nob_p, np_sharing_ratio, on=['year', 'category'])
new_nob_p.head()
```

```python
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")

sns.countplot(data = new_nob_p, x = 'award_portion', hue = 'prize_share', hue_order = ['1/1', '1/2', '1/3', '1/4'], palette = 'magma')

plt.title('Typical Prize Share distribution Overtime', fontsize=16, color = 'purple')
plt.xlabel('Prize Share', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

###### Answer

###### From the `Percentage proportion of award won and shared between laureates (winners) Overtime image above` shows how the nobel awards are shared (overtime)

###### From the `Typical Prize Share distribution Overtime image above` shows how the typiical share distribution is, which is :
###### 1-person awards consists of 1/1 prize share (solely to one person/organization)
###### 2-person awards consists of 1/2 prize share (shared between 2 people / organization)
###### 3-person awards consists of 1/2, 1/3 and 1/4 prize share (shared between 3 people / organization,
###### where one gets to keep 1/2 share while the remaining 2 shares half (1/4 each) or the 3 shares the award equally (1/3 each))
